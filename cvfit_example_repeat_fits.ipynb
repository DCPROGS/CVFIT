```python
%matplotlib inline
from pylab import*
```

```python
import numpy as np
from cvfit.equations import Hill 
from cvfit.data import XYDataSet
from cvfit.fitting import SingleFitSession
```

```python
conc = np.array([0.1, 0.3, 1.0, 3.0, 10.0]) # , 30.0, 100.0
sd = 0.1
res = []
for i in range(1000):
    eq = Hill('Hill', pars=[0.0, 1.0, 1.0, 1.0])
    resp = eq.calculate_random(conc, sd)
    dataset = XYDataSet()
    dataset.from_columns(conc, resp)
    #print (set)

    fs = SingleFitSession(dataset, eq)
    fs.fit()
    res.append(eq.pars)
results = np.array(res)
```

```python
def plot_hist(n, i, label, xl):
    subplot(2, 3, n)
    hist(results[:, i+1])
    xlabel(label)
    ylabel('Number')
    xlim(xl)

def plot_corr(n, i, j, l1, l2, xl, yl):
    subplot(2, 3, n)
    plot(results[:, i+1], results[:,j+1], 'b.')
    xlabel(l1)
    xlim(xl)
    ylabel(l2)
    ylim(yl)
    grid(True)
```

```python
plot_hist(1, 0, 'Ymax', (0,2))
plot_hist(2, 1, 'K', (0,7))
plot_hist(3, 2, 'nH', (0,7))
plot_corr(4, 0, 1, 'Ymax', 'K', (0,2), (0,7))
plot_corr(5, 0, 2, 'Ymax', 'nH', (0,2), (0,7))
plot_corr(6, 1, 2, 'K', 'nH', (0,7), (0,7))
```
