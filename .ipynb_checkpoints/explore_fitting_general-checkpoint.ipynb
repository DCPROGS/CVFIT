```python
%matplotlib inline
from pylab import*
```

```python
import math
from scipy.optimize import curve_fit
from scipy.special import binom
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

def popen(c, E, K, n):
    '''
    Mechanism of sequential binding of n molecules followed by opening.

    R <-> AR <-> A2R <-> ... <-> AnR <-> AnR*

    coeff : list [E, K, n]
    '''

    Eterm = E * c**n / K
    Kterm = 0.0
    for r in range(int(math.ceil(n))):
        Kterm += binom(int(math.ceil(n)), r) * (c / K)**r
    return  Eterm / (1 + Kterm + Eterm)
```

```python
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
np.random.seed(1729)
y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y + y_noise
```

```python
popt1, pcov = curve_fit(func, xdata, ydata)
popt1
```

```python
pcov
```

```python
popt2, pcov = curve_fit(func, xdata, ydata, bounds=(0, [3., 1., 0.5]))
popt2
```

```python
plot(xdata, ydata, 'b-', label='data')
plot(xdata, func(xdata, *popt1), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt1))
plot(xdata, func(xdata, *popt2), 'g--',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt2))
xlabel('x')
ylabel('y')
legend();
```
